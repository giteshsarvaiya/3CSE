<h4>17 Matplotlib</h4>
<h5>>  Matplotlib is a graph plotting library (module) in python that serves as a visualization utility.
<br/>> Matplotlib was created by John D. Hunter.
<br/>> Matplotlib is open source and we can use it freely.
<br/>> To install matplotlib we use: 
<br/>>>> pip install matplotlib
<br/>> To use matplotlib we use: 
<br/>>>> import matplotlib
<hr/>
Pyplot
<br/>> Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias: 
<br/>>>> import matplotlib.pyplot as plt
<hr/>
Setting figure size using: plt.figure(figsize=(x,y))
<br/>>>> example: plt.figure(figsize=(3,3)) # 3 inches by 3 inches
<br/>> figsize() takes two parameters- width and height (in inches). 
<br/>> By default the values for width and height are 6.4 and 4.8 respectively.
<hr/>
Plotting: using plot() function
<br/>>>> example: plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)
<br/>> After setting the figure size we call the plot() function.
<br/>> The plot() function is used to draw points (markers) in a diagram.
<br/>> By default, the plot() function draws a line from point to point.
<br/>> The function takes parameters for specifying points in the diagram.
<br/>>>> Parameter-1 (optional): is an array containing the points on the x-axis. This parameter is optional as it takes by default the values from 0 onwards depending upon the values in parameter-2
<br/>>>> Parameter-2 (compulsory): is an array containing the points on the y-axis.
<br/>> If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.
</h5>

```python
# Installing matplotlib using: pip install matplotlib
pip install matplotlib
```

```python
# importing matplotlib module(library)
import matplotlib
```

```python
# Using pyplot.plot() to plot a line 
# from position (1, 3) to position (8, 10)
# we can set the size of the plotted figure using:
# plt.figure(figsize=(row-inches,col-inches))
import matplotlib.pyplot as plt
x=[1,3] # we can use numpy arrays instead
y=[8,10] # we can use numpy arrays instead
plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.show()
```

<h4>(7) Multiple line plots in a single plot area</h4>
<h5>> We need to call the plot() function more than one time to draw more than one line charts in a single plot area.
<br/>> As we require to display a legend to denote which line represents which data so we need to use the 'label' parameter in the plot() function.
<br/>>>> plt.plot(x,y, color = 'value',label="value")
<br/>>>>>> the 'label' parameter is required when displaying legend
<br/>> We can call plot() function multiple times, that will cause the lines to be plotted in a single plot area.
<br/>> We call the legend() function after plot() to display the legend in the chart.
<br/>>>> plt.legend()
</h5>

```python
# WAP to plot line plots from "Open" and "Close" values
# present in "NIFTY50_EOD.csv" in the same chart, 
# use colour blue for "Open" # and colour red for "Close" values. 
# Provide legend in the figure.
# plt.plot(x,y,color,label)
# plt.legend()
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("NIFTY50_EOD.csv")
y1=df["Open"]
y2=df["Close"]
plt.plot(y1,color="Blue",label="Open")
plt.plot(y2,color="Red",label="Close")
plt.legend()
plt.show()
```

<h4>(8) (i) Subplots</h4>
<h5>> With the subplot() function we can draw multiple plots in one figure
<br/>>>> subplot(nrows, ncols, index) 
<br/>> The subplot() function takes three arguments that describes the layout of the figure.
<br/>> The layout is organized in rows and columns, which are represented by the first and second argument.
<br/>> The third argument represents the index of the current plot (index starts from 1 onwards).
<br/>> To avoid overlapping of plots in a subplots we use:
<br/>>>> plt.tight_layout()
</h5>

```python
# WAP to display separate line charts for "Open","High","Low" and "Close"
# values from "NIFTY50_EOD.csv" in a single figure.
# plt.subplot(nrows, ncols, index)
# plt.tight_layout()
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("NIFTY50_EOD.csv")
y1=df["Open"]
y2=df["High"]
y3=df["Low"]
y4=df["Close"]
# Plot-1
plt.subplot(2,2,1)
plt.plot(y1)
# Plot-2
plt.subplot(2,2,2)
plt.plot(y2)
# Plot-3
plt.subplot(2,2,3)
plt.plot(y3)
# Plot-4
plt.subplot(2,2,4)
plt.plot(y4)
plt.tight_layout() # auto adjust plots to avoid overlapping
plt.show()
```

<h4>(9) Scatter plots</h4>
<h5>> The scatter() function plots one dot for each observation. 
<br/>> It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:
<br/>>>> plt.scatter(x, y, color = 'value',label="value")
<br/>>>>>> the 'label' parameter is required when displaying legend
<br/>> We can call scatter() function multiple times, that will cause the points to be plotted in a single plot area.
</h5>

```python
# WAP to plot scatter plots from "Open" and "Close" values
# present in "NIFTY50_EOD.csv" in the same chart, 
# use colour blue for "Open"and colour red for "Close" values. 
# Provide legend in the figure.
# plt.scatter(x,y,color)
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("NIFTY50_EOD.csv")
x1=df["Open"]
x2=df["Close"]
y=df["Date"] # getting y-axis values from "Date" column
plt.scatter(x1,y,color="Blue",label="Open")
plt.scatter(x2,y,color="Red",label="Close")
plt.legend()
plt.show()
```
